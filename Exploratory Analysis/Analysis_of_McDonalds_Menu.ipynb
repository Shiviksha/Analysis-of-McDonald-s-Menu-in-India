###The box and whisker plot is used to understand variability in each category of the menu in terms of various nutrional values. The box represents 50% of the data values whereas the lower whisker represents the bottom 25% of the data and the upper whisker represents the top 25% of the data. The dots represent the outliers in each category. Hence, the box and whisker plot can also be used to visualize outliers in the data and clean and process the data accordingly.

```python
plot=sns.boxplot(x='Menu Category',y='Energy (kCal)',data=df)
plt.setp(plot.get_xticklabels(),rotation=45)
plot
```

```python
plot1=sns.boxplot(x='Menu Category', y='Total carbohydrate (g)', data=df)
plt.setp(plot1.get_xticklabels(), rotation=45)
plot1
```

Plotting the box and whisker plot for Trans fat gives an extreme value of outlier in the Regular Menu which makes vizualization of the boxplot difficult.
To remove the outlier we calculate interquartile range, estimate upper and lower limits, and then remove point lying beyond these values.


```python
Q1=df['Trans fat (g)'].quantile(0.25)
Q3=df['Trans fat (g)'].quantile(0.75)
IQR=Q3-Q1
UL=Q3+1.5*IQR
LL=Q1-1.5*IQR
upper_array=np.where(df['Trans fat (g)']>=UL)[0]
lower_array=np.where(df['Trans fat (g)']<=LL)[0]

indices_to_drop = np.concatenate((upper_array, lower_array))
df.drop(index=indices_to_drop, inplace=True)
```

```python
plot2=sns.boxplot(x='Menu Category', y='Trans fat (g)', data=df)
plt.setp(plot2.get_xticklabels(), rotation=45)
plot2
```

The heatmap is used to represent correlation between two variables with 1 signifying a perfect positive correlation, -1 signifying a perfect negative correlation and 0 signifying no correlation.

```python
heatmap=sns.heatmap(df.corr(), vmin=-1,vmax=1,cmap="PiYG",annot=True)
```

#Interpretation:

An average McDonald's value meal consists of 2 items from Regular Menu and 1 from Beverages Menu. One single meal will, on an average, constitute of approximately 880 KCals. If the meal is accompanied by a dessert then the average calories increase to 1042 Kcals.

This implies that a single meal constitutes of more than 50% of the daily calorie intake, 58% of the protein intake, 57% of Total fats of which 89% of Saturated fats, ~52% of carbohydrates and 65% of total sugar.

Additionally, The heatmap also shows that Energy (Kcals) in a meal is not always directly proportional to other nutrient values. Therefore calories in itself are not a sufficient indicator of a healthy meal.

To conclude: The numbers are indicative of an unbalanced meal with high contents of saturated fats and sugar. Since beverages menu has the highest average sugar, cutting it down would significantly reduce the intake and bring it down to 33.2%.

To maintain a healthy lifestyle and to restrict to an average daily diet requirements, the meals/items from the McDonald's menu must be consumed mindfully by checking all of their nutritional value and not consider Energy solely as an indicator of balanced nutritious meal.
